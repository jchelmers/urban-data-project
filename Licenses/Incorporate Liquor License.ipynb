## Load health inspection data, including latitude and longitude

```python
violations_df = pd.read_csv('../Geocoding-converting addresses into long/health_inspect_cleaned_loc_added.csv', index_col=0)
violations_df = violations_df.reindex(sorted(violations_df.index.values)) # Sort by CAMIS
display(violations_df)
```

```python
threshold = 0.0001 # We'll count lat/long values as a match if they're within this threshold

violations_df['Liquor License'] = ['NA']*len(violations_df)

for indx, r in violations_df.iterrows():
    lat = violations_df.ix[indx,'latitude']
    lng = violations_df.ix[indx,'longitude']
    # Find matching latitude(s)
    mask = abs(liquor['Latitude']-lat)<=threshold
    matching_lat = liquor.ix[mask, :]
    if len(matching_lat) > 1: # Check longitude(s) if we've found matching latitude(s)
        mask = abs(matching_lat['Longitude']-lng)<=threshold
        matching_latlong = matching_lat.ix[mask, :]
    if len(matching_latlong)==1: # Only add to dataframe if we've found one and only one match
        violations_df.ix[indx,'Liquor License'] = str(matching_latlong.ix[:,'License Type Code'].values)

display(violations_df)
```
