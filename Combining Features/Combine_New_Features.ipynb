```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
from sklearn import preprocessing
from sklearn import linear_model
from sklearn import svm
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import BernoulliNB
from sklearn.ensemble import RandomForestClassifier
%matplotlib inline
```

```python
#read all
rating = pd.read_csv('../Ratings/Health_Insp_weather_loc_rating.csv', index_col=0) # change  
weather = pd.read_csv('../Weather/health_inspect_cleaned_weather_added.csv', index_col=0) # change  
license = pd.read_csv('../Licenses/health_inspect_loc_license.csv', index_col=0) # change  
complaints = pd.read_csv('../311 Complaints/vdf_with_complaints.csv', index_col=0) # change   
prior_vs = pd.read_csv('../prior_violations.csv', index_col = 0)
complaints
```

```python
rating = rating[['rating', 'rating_null', 'rating_null_zip']]
#weather = weather[['3-day temp', '3-day humidity']]
complaints = complaints[['food_poisoning_complaints',
       'food_establishment_complaints', 'electric_complaints',
       'safety_complaints', 'rodent_complaints', 'dirty_conditions_complaints',
       'general_complaints', 'missed_collection_complaints',
       'police_matter_complaints']]

license['liq_license'] = ~license['Liquor License'].isnull()
license['side_license'] = ~license['Sidewalk License'].isnull()
license = license[['liq_license', 'side_license']]
```

```python
extra_info_df = weather
extra_info_df = extra_info_df.merge(prior_vs, left_index = True, right_index=True)
extra_info_df = extra_info_df.merge(rating, left_index = True, right_index=True)
extra_info_df = extra_info_df.merge(complaints, left_index = True, right_index=True)
extra_info_df = extra_info_df.merge(license, left_index = True, right_index=True)
```

```python
#write in a file
extra_info_df.to_csv('../violations_all.csv')
```
